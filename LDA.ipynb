```python
import pandas as pd

import gensim
import gensim.corpora as corpora
from gensim.models import CoherenceModel

from pprint import pprint

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python
data = pd.read_csv('Greco cleaned.csv')
```

```python
## tokenization
baskets = []
for id in data['ID'].unique():
    for tp in data.loc[data['ID'] == id, 'TP-Date'].unique():
        for rec in data.loc[(data['ID'] == id) & (data['TP-Date'] == tp), 'RecNum'].unique():
            baskets.append(data.loc[(data['ID'] == id) & (data['TP-Date'] == tp) & (data['RecNum'] == rec), 'Item'].tolist())
            
id2word = corpora.Dictionary(baskets)
corpus = [id2word.doc2bow(b) for b in baskets]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=id2word,num_topics=35,passes=1000)
#pprint(lda_model.print_topics(25))

print('Perplexity: ', lda_model.log_perplexity(corpus))

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=baskets, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)

# Visualize the topics
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, id2word)
vis
```
